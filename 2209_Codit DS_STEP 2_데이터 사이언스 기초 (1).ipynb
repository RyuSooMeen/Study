### 6) 연속된 값들이 담긴 numpy array 생성 : np.arange

arange 함수는 파이썬의 기본 함수인 range와 굉장히 비슷한 원리로 동작합니다.

#### ① 파라미터 1개
#### arange(m)을 하면 0부터 m-1까지의 값들이 담긴 numpy array가 리턴됩니다.

```python
array1 = np.arange(6)
print(array1)
```

#### ② 파라미터 2개
#### arange(n, m)을 하면 n부터 m-1까지의 값들이 담긴 numpy array가 리턴됩니다.

```python
array1 = np.arange(2, 7)
print(array1)
```

#### ③ 파라미터 3개
#### arange(n, m, s)를 하면 n부터 m-1까지의 값들 중 간격이 s인 값들이 담긴 numpy array가 리턴됩니다.

```python
array1 = np.arange(3, 17, 3)
print(array1)
```

## 14. numpy 기본 통계

### 1) 최댓값, 최소값 : max(),min()

```python
import numpy as np

array1 = np.array([14, 6, 13, 21, 23, 31, 9, 5])

print(array1.max()) # 최댓값
print(array1.min()) # 최솟값
```

```python
import numpy as np

array1 = np.array([14, 6, 13, 21, 23, 31, 9, 5])

print(array1.mean()) # 평균값
```

### 2) 중앙값 : np.median
#### median 메소드를 사용하면 중간값을 구할 수 있는데요. 특이하게 median은 numpy array의 메소드가 아니라 numpy의 메소드입니다.

```python
import numpy as np

array1 = np.array([8, 12, 9, 15, 16])
array2 = np.array([14, 6, 13, 21, 23, 31, 9, 5])

print(np.median(array1)) # 중앙값
print(np.median(array2)) # 중앙값
```

```python
import numpy as np

array1 = np.array([14, 6, 13, 21, 23, 31, 9, 5])

print(array1.std()) # 표준 편차
print(array1.var()) # 분산
```

## 04. DataFrame을 만드는 다양한 방법

### 1) From list of lists, array of arrays, list of series

```python
import numpy as np
import pandas as pd

two_dimensional_list = [['dongwook', 50, 86], ['sineui', 89, 31], ['ikjoong', 68, 91], ['yoonsoo', 88, 75]]
two_dimensional_array = np.array(two_dimensional_list)
list_of_series = [
    pd.Series(['dongwook', 50, 86]), 
    pd.Series(['sineui', 89, 31]), 
    pd.Series(['ikjoong', 68, 91]), 
    pd.Series(['yoonsoo', 88, 75])
]

# 아래 셋은 모두 동일합니다
df1 = pd.DataFrame(two_dimensional_list)
df2 = pd.DataFrame(two_dimensional_array)
df3 = pd.DataFrame(list_of_series)

df1
```

### 2) From dict of lists, dict of arrays, dict of series

#### 파이썬 사전(dictionary)으로도 DataFrame을 만들 수 있습니다.
#### 사전의 key로는 column 이름을 쓰고, 그 column에 해당하는 리스트, numpy array, 혹은 pandas Series를 사전의 value로 넣어주면 됩니다.

```python
import numpy as np
import pandas as pd

names = ['dongwook', 'sineui', 'ikjoong', 'yoonsoo']
english_scores = [50, 89, 68, 88]
math_scores = [86, 31, 91, 75]

dict1 = {
    'name': names, 
    'english_score': english_scores, 
    'math_score': math_scores
}

dict2 = {
    'name': np.array(names), 
    'english_score': np.array(english_scores), 
    'math_score': np.array(math_scores)
}

dict3 = {
    'name': pd.Series(names), 
    'english_score': pd.Series(english_scores), 
    'math_score': pd.Series(math_scores)
}


# 아래 셋은 모두 동일합니다
df1 = pd.DataFrame(dict1)
df2 = pd.DataFrame(dict2)
df3 = pd.DataFrame(dict3)

df1
```

### 3) From list of dicts

#### 리스트가 담긴 사전이 아니라, 사전이 담긴 리스트로도 DataFrame을 만들 수 있습니다.

```python
import numpy as np
import pandas as pd

my_list = [
    {'name': 'dongwook', 'english_score': 50, 'math_score': 86},
    {'name': 'sineui', 'english_score': 89, 'math_score': 31},
    {'name': 'ikjoong', 'english_score': 68, 'math_score': 91},
    {'name': 'yoonsoo', 'english_score': 88, 'math_score': 75}
]

df = pd.DataFrame(my_list)
df
```

## 07. 서류 전형 합격 여부

### 실습과제
토익 시험은 LC(듣기) 파트와 RC(독해) 파트로 이루어져 있습니다. 각 파트가 495점 만점, 총 990점이 만점입니다.

#### “서울 항공”에 입사하기 위해서는 토익 점수를 제출해야 하는데요. 각 파트가 최소 250점, 총 점수가 최소 600점이 되어야 서류 전형을 합격할 수 있습니다.

#### 기존 DataFrame에 “합격 여부”라는 column을 추가하고, 합격한 지원자는 불린 값 True, 불합격한 지원자는 불린 값 False를 넣어주면 됩니다.

```python
import pandas as pd
    
df = pd.read_csv('data/toeic.csv')

pass_total = df['LC'] + df["RC"] > 600
pass_both = (df['LC'] >= 250) & (df["RC"] >= 250)
df['합격여부'] = pass_total & pass_both
#df.loc[(df['LC']+df['RC'] >= 600) & (df['LC'] >= 250) & (df['RC'] >= 250), '합격여부'] = True
#df.loc[(df['LC']+df['RC'] < 600) | (df['LC'] < 250) | (df['RC'] < 250) ,'합격여부'] = False
df
```

#### 5. 폐강 등의 이유로 status가 “not allowed”인 수강생은 room assignment 또한 “not assigned”가 되어야 합니다.


```python
import pandas as pd

df = pd.read_csv('data/enrolment_2.csv')
df
```

```python
df['room assignment'] = 'not assigned'
cn = df['course name'].value_counts()
cn_A = list(cn[cn>=80].index)
cn_L = list(cn[(cn<80) & (cn>=40)].index)
cn_M = list(cn[(cn<40) & (cn>=15)].index)
cn_S = list(cn[(cn<15) & (cn>=5)].index)
for i in range(len(df)):
    if df.iloc[i,2] in cn_A:
        df.iloc[i,4] = 'Auditorium'
    elif df.iloc[i,2] in cn_L:
        df.iloc[i,4] = 'Large room'
    elif df.iloc[i,2] in cn_M:
        df.iloc[i,4] = 'Medium room'
    elif df.iloc[i,2] in cn_S:
        df.iloc[i,4] = 'Small room'
        
df.loc[df['status'] == 'not allowed','room assignment'] = 'not assigned'
df
```

```python
import pandas as pd

df = pd.read_csv('data/enrolment_3.csv')

df
```

```python
ra = list(df['room assignment'].unique())
A = list(df.loc[df['room assignment'] == ra[1],'course name'].unique())
A.sort()
S = list(df.loc[df['room assignment'] == ra[2],'course name'].unique())
S.sort()
M = list(df.loc[df['room assignment'] == ra[3],'course name'].unique())
M.sort()
L = list(df.loc[df['room assignment'] == ra[4],'course name'].unique())
L.sort()
for i in range(len(df)):
    if df.iloc[i,4] == ra[1]:
        s = ra[1]+"-"+str(A.index(df.iloc[i,2]) + 1)
        df.iloc[i,4] = s
    elif df.iloc[i,4] == ra[2]:
        s = "Small-"+str(S.index(df.iloc[i,2]) + 1)
        df.iloc[i,4] = s
    elif df.iloc[i,4] == ra[3]:
        s = "Medium-"+str(M.index(df.iloc[i,2]) + 1)
        df.iloc[i,4] = s
    elif df.iloc[i,4] == ra[4]:
        s = "Large-"+str(L.index(df.iloc[i,2]) + 1)
        df.iloc[i,4] = s

df.rename(columns={'room assignment':'room number'}, inplace = True)
df
```
