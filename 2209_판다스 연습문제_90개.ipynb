### Question 57
#### Income_Categoryì˜ ì¹´í…Œê³ ë¦¬ë¥¼ map í•¨ìˆ˜ë¥¼ ì´ìš©í•˜ì—¬ ë‹¤ìŒê³¼ ê°™ì´ ë³€ê²½í•˜ì—¬ newIncome ì»¬ëŸ¼ì— ë§¤í•‘í•˜ë¼ 

```python
pd.DataFrame({"ê¸°ì¤€": ["Less than $40K : a", "$40K - $60K : b", "$60K - $80K : c","$80K - $120K : d","$120K +: e"]})
```

```python
df.loc[df["Income_Category"] == "Unknown", "newIncome"] = "N"
df.loc[df["Income_Category"] == 'Less than $40K', "newIncome"] = "a"
df.loc[df["Income_Category"] == '$40K - $60K', "newIncome"] = "b"
df.loc[df["Income_Category"] == '$60K - $80K', "newIncome"] = "c"
df.loc[df["Income_Category"] == '$80K - $120K', "newIncome"] = "d"
df.loc[df["Income_Category"] == '$120K +', "newIncome"] = "e"
df.newIncome
```

```python
dic = {
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e'   
}
df['newIncome'] = df.Income_Category.map(lambda x : dic[x])
df.newIncome
```

### â— Question 58

#### Income_Categoryì˜ ì¹´í…Œê³ ë¦¬ë¥¼ apply í•¨ìˆ˜ë¥¼ ì´ìš©í•˜ì—¬ ë‹¤ìŒê³¼ ê°™ì´ ë³€ê²½í•˜ì—¬ newIncome ì»¬ëŸ¼ì— ë§¤í•‘í•˜ë¼

```python
def changeCategory(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':   
        return 'b'
    elif x =='$60K - $80K':    
        return 'c'
    elif x =='$80K - $120K':   
        return 'd'
    elif x =='$120K +' :     
        return 'e'

df['newIncome']  =df.Income_Category.apply(changeCategory)
df.newIncome
```

```python
def change_date(x):
    import datetime
    hour = x.split(":")[1]
    date = x.split(":")[0]
    
    if hour == '24':
        hour = '00:00:00'
        FinalDate = pd.to_datetime(date + " "+hour) + datetime.timedelta(days = 1)
    else:
        hour = hour + ":00:00"
        FinalDate = pd.to_datetime(date+" "+hour)
    return FinalDate
df["(ë…„-ì›”-ì¼:ì‹œ)"] = df["(ë…„-ì›”-ì¼:ì‹œ)"].apply(change_date)
df
```
